```python

!which python
import matplotlib.pyplot as plt

# Import functions
from stencil2d_baseline import calculations as baseline_calc
from stencil2d_numpy import calculations as numpy_calc
from stencil2d_torch import calculations as torch_calc
from stencil2d_jax_base import calculations as jax_calc
```

```python
functions = [baseline_calc, jax_calc, torch_calc]
function_colors = ["r", "b", "g"]
function_names=["numpy", "jax", "torch"]
range_precision = ["32"]
nx = 64
ny_range = [16, 32, 48, 64, 80, 96, 112, 128]
nz = 128
num_iter = 10
num_halo = 2
```

```python
plt.figure()
times = []
for i, f in enumerate(functions):
    for ny in ny_range:
        for precision in range_precision:
            times.append(f(nx, ny, nz, num_iter, num_halo, precision, return_time=True))
    plt.plot(ny_range, times, function_colors[i], label=function_names[i])
    times = []
plt.legend()
plt.xlabel("ny")
plt.ylabel("runtime [s]")
plt.show()
```
